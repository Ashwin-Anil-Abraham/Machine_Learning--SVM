```python
import pandas as pd
sd=pd.read_csv("/content/Cancer_Data.csv")
print(sd.head())
```

```python
sd['diagnosis'].value_counts()          #"value count is used to know what are the values inside a column"
```

```python
sd.drop(['Unnamed: 32','id'],axis=1,inplace=True)
sd.head()
```

```python
sd.isna().sum()
```

```python
q=sd.iloc[:,1:]
#q
w=sd.iloc[:,0]
w
```

```python
#standard scalar:= used to scale down the values eqn=x-mean(x)/sd

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
#sc.fit(q)
#q=sc.transform(q)
#q
#OR FOR step 5 and 6
q=sc.fit_transform(q)
q
```

```python
from sklearn.model_selection import train_test_split
q_train,q_test,w_train,w_test=train_test_split(q,w,test_size=.2,random_state=42)
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.neighbors import KNeighborsClassifier    #we use knn just to compare the accuracy no need to do
from sklearn.svm import SVC
knn=KNeighborsClassifier(n_neighbors=5)
sv1=SVC(kernel='linear')
sv2=SVC(kernel='poly')
sv3=SVC(kernel='rbf')
#sv1.fit(q_train,w_train)
#w_predict1=sv1.predict(w_test)
#OR USING FOR LOOP for 10 and 11 step


l=[sv1,sv2,sv3,knn]
for i in l:
  i.fit(q_train,w_train)
  w_predict=i.predict(q_test)
  print(accuracy_score(w_test,w_predict))
  print(confusion_matrix(w_test,w_predict))
  print(classification_report(w_test,w_predict))
```
